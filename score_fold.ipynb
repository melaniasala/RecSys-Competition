```python
import numpy as np
import pandas as pd
```

```python
n_folds = 5
fold_scores = {}
fold_rankings = {}
```

```python
for fold in range(n_folds):
    scores_name = "xgboost_scores_fold" + str(fold + 1)

    print("Loading scores for fold", fold, "...")
    # TODO: modify path
    scores = np.load('C:\\Users\\melan\\shared-folder\\RecSys\\RecSys-Competition\\scores_xgbfold\\' + scores_name + '.npy')

    # Assign the variable to the dictionary 
    fold_scores["scores_fold" + str(fold + 1)] = scores
```

```python
for fold in range(n_folds):
    rankings_name = "xgboost_ranking_fold" + str(fold + 1)

    print("Loading rankings for fold", fold, "...")
    # TODO: modify path
    rankings = np.load('C:\\Users\\melan\\shared-folder\\RecSys\\RecSys-Competition\\scores_xgbfold\\' + rankings_name + '.npy')

    # Assign the variable to the dictionary 
    fold_rankings["rankings_fold" + str(fold + 1)] = rankings
```

```python
# Consistency check!
for fold in range(n_folds):
    rankings_name = "rankings_fold" + str(fold + 1)

    print(fold_rankings[rankings_name])
```

```python
items_selected = fold_rankings["rankings_fold1"]
```

```python
fold_scores["scores_fold1"]
```

```python
# For each Recommender object we compute the scores and store them in scores_batch_array
scores_batch_array = np.array([fold_scores["scores_fold" + str(fold + 1)] for fold in range(n_folds)])

# Now compute the scores by  averaging over the scores of each fold
scores = np.average(scores_batch_array, axis= 0)
```

```python
import tqdm

predictions = []

for user_id in range(len(scores)):
    scores_user = scores[user_id]
    items_user = items_selected[user_id]

    sorted_ranking = np.argsort(-scores_user)[:10]
    prediction = items_user[sorted_ranking]

    predictions.append(prediction) 
```

```python
predictions
```

```python
predictions_list = []
predictions_list = predictions

prediction_df = pd.read_csv('C:\\Users\\melan\\shared-folder\\RecSys\\RecSys-Competition\\data_target_users_test.csv', sep= ",",
                            header=0, 
                            dtype={0:int},
                            engine='python')
users = np.array(prediction_df['user_id'])
predictions_list = [predictions_list[i] for i in users]
truncated_recommendations = [inner_list[:10] for inner_list in predictions_list]
prediction_df['item_list'] = truncated_recommendations
def transform_items_to_string(item_list):
    return ' '.join(map(str, item_list))

prediction_df['item_list'] = prediction_df['item_list'].apply(transform_items_to_string)
print(prediction_df.head(10))
prediction_df.to_csv('submission.csv',index=False)
```
